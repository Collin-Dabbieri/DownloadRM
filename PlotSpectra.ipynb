```python
import numpy as np
import glob
from matplotlib import pyplot as plt
import matplotlib
import os
```

```python
#os.system('ls '+test_dir+'*_unred_z.npy')
print(test_dir+'*_unred_z.npy')
spectra_paths=glob.glob(test_dir+'*_unred_z.npy')
print(spectra_paths[:5])
print(len(spectra_paths))
```

```python
print(spectra_paths[0].split('-')[2])
spectra_mjds=[int(i.split('-')[2]) for i in spectra_paths]

idx_sort=np.argsort(spectra_mjds)
spectra_paths_sorted=np.array(spectra_paths)[idx_sort]
print(spectra_paths_sorted)
```

```python
cmap = matplotlib.colormaps.get_cmap('viridis')

for i in range(len(spectra_paths_sorted)):
    fraction=i/len(spectra_paths_sorted)
    data=np.load(spectra_paths_sorted[i],allow_pickle=True)
    WL=data.item().get('WL')
    INT=data.item().get('INT')
    ERR=data.item().get('ERR')

    plt.plot(WL,INT,c=cmap(fraction),alpha=0.1)
plt.axis([3500,5500,0,35])
plt.axvline(4980,c='r')
plt.axvline(5030,c='r')
```
